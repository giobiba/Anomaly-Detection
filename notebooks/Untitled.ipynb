```python
import numpy as np
from numpy.random import default_rng
from matplotlib import pyplot as plt
import PIL.Image as Image
import ImShow as I
from keras.datasets import mnist
import pickle
import pandas as pd
```

```python
with open('anomalydet_with_corruption.pickle', 'rb') as handle:
    saved = pickle.load(handle)
```

```python
corruptions = [0, 50, 200]
fig, ax = plt.subplots(2, len(corruptions), sharey=True)
plt.figure(figsize=(2, 3), dpi=300)
for c in corruptions:
    idx=saved[c]['all']['fscore'].index(max(saved[c]['all']['fscore']))
    rec = saved[c]['all']['recovered_x'][idx]
    cor = saved[c]['corrupted_x']
    lambdas_ = saved[c]['all']['lambdas']
    
    ax[0][corruptions.index(c)].imshow(I.tile_raster_images(X=rec, img_shape=(28,28), tile_shape=(10, 10),tile_spacing=(1, 1)), cmap='gray')
    ax[1][corruptions.index(c)].imshow(I.tile_raster_images(X=cor, img_shape=(28,28), tile_shape=(10, 10),tile_spacing=(1, 1)), cmap='gray')   

fig.savefig('smth.png', dpi=500)
```
